#Importaciones

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

#Carga del csv

```python
df = pd.read_csv('df_anscombe.csv')
df.head()
```

#Descripcion de los valores del dataset

```python
df.groupby('group').describe()
```

#Grafico de la figuras

```python
# Crear figura y ejes
fig, axs = plt.subplots(2, 2, figsize=(10, 8))  # 4 subplots (2x2)
axs = axs.flatten()  # Para iterar fcilmente

# Grficos por grupo
for i, group in enumerate(df['group'].unique()):
    subset = df[df['group'] == group]
    axs[i].scatter(subset['x'], subset['y'], color='red')
    axs[i].set_title(f'Grupo {group}')
    axs[i].set_xlabel('x')
    axs[i].set_ylabel('y')
    axs[i].grid(True)

plt.tight_layout()
plt.show()
```

```python
# Sacar la regresion lineal en una sola grafica de todas los grupos
# Crear el grfico
plt.figure(figsize=(10, 6))

# Usar seaborn para trazar los puntos y la regresin
sns.lmplot(
    data=df,
    x='x',
    y='y',
    hue='group',
    markers=['o', 'o', 'o', 'o'],
    height=6,
    aspect=1.5,
    ci=None,
    palette='Set2'
)

plt.title('Datos de Anscombe con regresin lineal por grupo')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Definir colores para cada grupo
colores = {1: 'yellow', 2: 'red', 3: 'green', 4: 'violet'}

plt.figure(figsize=(10, 8))

# Graficar los puntos de cada grupo con un color distinto
for grupo, color in colores.items():
    data = df[df['group'] == grupo]
    plt.scatter(data['x'], data['y'], color=color, label=f'Grupo {grupo}', s=50)

# Calcular la lnea de regresin global (para todos los datos)
slope, intercept = np.polyfit(df['x'], df['y'], 1)
x_vals = np.array([df['x'].min(), df['x'].max()])
y_vals = intercept + slope * x_vals
plt.plot(x_vals, y_vals, color='red', lw=2, label='Lnea de regresin')

# Personalizacin del grfico
plt.xlabel('x')
plt.ylabel('y')
plt.title("Anscombe's Quartet con lnea de regresin global")
plt.grid(True)
plt.legend()
plt.show()
```

#Margen de error

```python
# Colores por grupo
colores = {1: 'yellow', 2: 'red', 3: 'green', 4: 'violet'}

plt.figure(figsize=(10, 8))

# Graficar los puntos por grupo
for grupo, color in colores.items():
    data = df[df['group'] == grupo]
    plt.scatter(data['x'], data['y'], color=color, label=f'Grupo {grupo}', s=50)

# Usar seaborn para trazar lnea de regresin con banda de confianza del 95%
sns.regplot(x='x', y='y', data=df, scatter=False, ci=95, color='blue', line_kws={'label':'Lnea de regresin'})

# Personalizacin del grfico
plt.xlabel('x')
plt.ylabel('y')
plt.title("Anscombe's Quartet con lnea de regresin global y banda de confianza")
plt.grid(True)
plt.legend()
plt.show()
```
